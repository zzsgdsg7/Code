```python
import matplotlib.pyplot as plt  # 导入科学绘图包
import networkx as nx
import numpy as np
import csv
```

```python
node_path = r'C:\Users\zzsgdsg7\Desktop\netnode.csv'
edge_path = r'C:\Users\zzsgdsg7\Desktop\netedge.csv'
```

```python
with open (node_path) as nodecsv:
    nodereader = csv.reader(nodecsv)
    nodes = [n for n in nodereader][1:]
node_names = [n[0] for n in nodes]

with open(edge_path) as edgecsv: # Open the file
    edgereader = csv.reader(edgecsv) # Read the csv
    edges = [tuple(e) for e in edgereader][1:] # Retrieve the data
```

```python
G = nx.Graph()
G.add_nodes_from(node_names)
G.add_edges_from(edges)
print(G)
```

```python
degree = nx.degree_histogram(G)  # 返回图中所有节点的度分布序列
```

```python
print(degree)
```

```python
#可视化
x = range(len(degree))  # 生成X轴序列，从1到最大度
#y = [z / float(sum(degree)) for z in degree]
y = [z  for z in degree]# 将频次转化为频率
plt.figure(figsize=(5.8, 5.2), dpi=150)
plt.rcParams['font.sans-serif'] = ['SimHei']  # 用来正常显示中文标签
plt.xlabel("Degree", size=14)  # Degree
plt.ylabel("Frequency", size=14)  # Frequency
plt.xticks(fontproperties='Times New Roman', size=13)
plt.yticks(fontproperties='Times New Roman', size=13)
plt.loglog(x, y, '.')
plt.show()  # 显示图表
```
