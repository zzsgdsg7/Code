```python
def knn_mean(ts, n):#KNN方法处理数据
    out = np.copy(ts)
    for i, val in enumerate(ts):
        if np.isnan(val):#判断值是否是空
            n_by_2 = np.ceil(n/2)
            lower = np.max([0, int(i-n_by_2)])
            upper = np.min([len(ts)+1, int(i+n_by_2)])
            ts_near = np.concatenate([ts[lower:i], ts[i:upper]])
            out[i] = np.nanmean(ts_near)
    return out
```

```python
for indexs in AQI.columns:
    if indexs =='time':
        continue
    col_null = col_null = AQI[indexs].isnull().sum(axis = 0)
    if col_null < 1100:
        AQI[indexs] = knn_mean(AQI[indexs].values,24)
    if col_null > 1100:
        AQI = AQI.drop(columns = indexs)    
```
